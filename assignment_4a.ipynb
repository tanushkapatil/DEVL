```python
import matplotlib.pyplot as plt
import itertools
columns = ['CRIM','ZN', 'INDUS', 'CHAS', 'NOX']

# Create a figure with subplots
fig, axes = plt.subplots(nrows=len(columns), ncols=len(columns), figsize=(12, 12))

# Loop through each combination of columns
for i, (col1, col2) in enumerate(itertools.product(columns, repeat=2)):
    ax = axes[i // len(columns), i % len(columns)]
    ax.scatter(df[col1], df[col2], color='cyan')
    ax.set_xlabel(col1)
    ax.set_ylabel(col2)
    ax.set_title(f'{col1} vs {col2}')
    ax.grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import numpy as np
# Check for NaN values
# print(df.isna().sum())

# Option 1: Remove rows with NaN values
# df_cleaned = df.dropna()

# Option 2: Replace NaN with a default value (e.g., 0 or the mean of the column)
df = df.fillna(df.median())

# Option 3: Check for infinite values and replace them
# df.replace([np.inf, -np.inf], np.nan, inplace=True)
# df_cleaned = df.dropna()
df
df.isnull().sum()
```

```python
import sklearn
from sklearn.linear_model
import LinearRegression

X = df.iloc[:,:-1]
Y = df.iloc[:,-1]
lr = LinearRegression()
lr.fit(X, Y)
X
```

```python
Y
```

```python
print(X.shape)
print(Y.shape)
```

```python
from sklearn.model_selection import train_test_split

# Assuming X and Y are already defined
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print("Mean Square Error : ", mse)
```

```python
plt.scatter(y_test, y_pred, c = 'green')
plt.xlabel("Price: in $1000's")
plt.ylabel("Predicted value")
plt.title("True value vs predicted value : Linear Regression")
plt.show()
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, y_pred)
print(score)
```
