```python
import requests
from bs4 import BeautifulSoup
URL = "https://www.geeksforgeeks.org/python-web-scraping-tutorial/"
r = requests.get(URL)
soup = BeautifulSoup(r.content , 'html5lib')
print(soup.prettify)
```

```python
titles = soup.find_all('h3')
for title in titles:
    print(title.text)
```

```python
links = soup.find_all("a")
print(links)
```
