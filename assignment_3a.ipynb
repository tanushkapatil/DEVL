```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating the dataset as a DataFrame
data_dict = {
    'company': ['alfa-romero', 'alfa-romero', 'alfa-romero', 'audi', 'audi',
                'volkswagen', 'volkswagen', 'volvo', 'volvo'],
    'body-style': ['convertible', 'convertible', 'hatchback', 'sedan', 'sedan',
                   'sedan', 'sedan', 'sedan', 'wagon'],
    'engine-type': ['dohc', 'dohc', 'ohcv', 'ohc', 'ohc',
                    'ohc', 'ohc', 'ohc', 'ohc'],
    'num-of-cylinders': ['four', 'four', 'six', 'four', 'five',
                         'four', 'four', 'four', 'four'],
    'horsepower': [111, 111, 154, 102, 115, 85, 52, 114, 114],
    'average-mileage': [21, 21, 19, 24, 18, 27, 37, 23, 23],
    'price': [13495.0, 16500.0, 16500.0, 13950.0, 17450.0,
              7975.0, 7995.0, 12940.0, 13415.0]
}

# Convert to DataFrame
df = pd.DataFrame(data_dict)

# Creating box plots for horsepower, average mileage, and price
plt.figure(figsize=(15, 5))

# Box plot for horsepower
plt.subplot(1, 3, 1)
plt.boxplot(df['horsepower'], vert=True)
plt.title('Box Plot of Horsepower')
plt.ylabel('Horsepower')

# Box plot for average mileage
plt.subplot(1, 3, 2)
plt.boxplot(df['average-mileage'], vert=True)
plt.title('Box Plot of Average Mileage')
plt.ylabel('Average Mileage')

# Box plot for price
plt.subplot(1, 3, 3)
plt.boxplot(df['price'], vert=True)
plt.title('Box Plot of Price')
plt.ylabel('Price')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Creating dataset
np.random.seed(10)
data = np.random.normal(100, 20, 200)

# Detecting and removing outliers using IQR
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Define the bounds for non-outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the data to remove outliers
filtered_data = data[(data >= lower_bound) & (data <= upper_bound)]

# Plotting the boxplot without outliers
fig = plt.figure(figsize=(10, 7))
plt.boxplot(filtered_data)
plt.title("Boxplot after removing outliers")
plt.show()
```
