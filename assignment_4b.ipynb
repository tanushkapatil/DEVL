```python
mean_value = df['Insulin'][df['Insulin'] != 0].mode()[0]

df['Insulin'] = df['Insulin'].replace(0, mean_value)
df
```

```python
columns_to_fill = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

for column in columns_to_fill:
    if df[column].dtype in ['int64', 'float64']:
        mean_value = df[column][df[column] != 0].mean()
        df[column] = df[column].replace(0, mean_value)

print(df)
```
